```python

#to import the random library from python
import random

#to set the limits of each game level
easy_limit= 10
mod_limit=100
hard_limit=1000

# to keep track of scores
loss=0
win=0

def display_title():
    print("Guess the number!")
    print()

def easy_game():    
    number = random.randint(1, easy_limit)
    print("I'm thinking of a number from 1 to " + str(easy_limit) + "\n")
    count=1

    #to be able to use win/loss global variable inside the function
    global win
    global loss

    while count<=5:
        guess = int(input("Your guess: "))
        if guess < number:
            print("Too low.")
            count += 1

        elif guess > number:
            print("Too high.")
            count += 1

        elif guess == number:
            print("You guessed it in " + str(count) + " tries.\n")

            #to count wins & update scoreboard
            win+=1
            print("Your score is:", win, "wins", loss,"losses")
            return
    else:
        print("You lose! You passed 5 tries")

        #to count losses & update scoreboard
        loss+=1
        print("Your score is:", win, "wins", loss,"losses")




def mod_game():    
    number = random.randint(1, mod_limit)
    print("I'm thinking of a number from 1 to " + str(mod_limit) + "\n")
    count=1

    #to be able to use win/loss global variable inside the function
    global win
    global loss

    while count<=8:
        guess = int(input("Your guess: "))
        if guess < number:
            print("Too low.")
            count += 1

        elif guess > number:
            print("Too high.")
            count += 1

        elif guess == number:
            print("You guessed it in " + str(count) + " tries.\n")

            #to count wins & update scoreboard
            win+=1
            print("Your score is:", win, "wins", loss,"losses")
            return
    else:
        print("You lose! You passed 8 tries")

        #to count losses & update scoreboard
        loss+=1
        print("Your score is:", win, "wins", loss,"losses")



def hard_game():    
    number = random.randint(1, hard_limit)
    print("I'm thinking of a number from 1 to " + str(hard_limit) + "\n")
    count=1

    #to be able to use win/loss global variable inside the function
    global win
    global loss

    while count<=10:
        guess = int(input("Your guess: "))
        if guess < number:
            print("Too low.")
            count += 1

        elif guess > number:
            print("Too high.")
            count += 1

        elif guess == number:
            print("You guessed it in " + str(count) + " tries.\n")

            #to count wins & update scoreboard
            win+=1
            print("Your score is:", win, "wins", loss,"losses")
            return
    else:
        print("You lose! You passed 10 tries")

        #to count losses & update scoreboard
        loss+=1
        print("Your score is:", win, "wins", loss,"losses")
        return

#function for printing name
def name ():
    name=input("Please enter your name: ")
    print("Welcome, ", name)


def main():
    #using the name printing function
    name()

    #using the title function
    display_title()

    
    #a while loop to take user input whether they'd like to continue or not
    again = "y"
    while again.lower() == "y":

        #ask user about level wanted to play & will be printed after each round
        level=int(input("choose 1) Easy, 2) Moderate, 3) Hard"))

        #an if loop that will use the game function based on the level chosen
        if level==1:
            easy_game()
            again = input("Would you like to play again? (y/n): ")
            print()

        elif level==2:
            mod_game()
            again = input("Would you like to play again? (y/n): ")
            print()

        elif level==3:
            hard_game()
            again = input("Would you like to play again? (y/n): ")
            print()
    
    print("Bye!")

# if started as the main module, call the main function
if __name__ == "__main__":
    main()

```
